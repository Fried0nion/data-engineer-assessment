```python

import re
import pandas as pd

col = next((c for c in DataDF.columns if c.lower() == 'alamat4'), None)
if col is None:
    print("Column 'Alamat4' not found.")
else:
    logs = []
    new_vals = []
    for idx, val in DataDF[col].items():
        if pd.isna(val):
            logs.append(f"Row {idx}: NaN -> NaN (skipped)")
            new_vals.append(pd.NA)
            continue
        s = str(val).strip()
        if re.search(r',\s*KOTA\b', s, flags=re.IGNORECASE):
            cleaned = re.sub(r',\s*KOTA\b', '', s, flags=re.IGNORECASE).strip()
            new = 'KOTA ' + cleaned
            logs.append(f"Row {idx}: '{s}' -> '{new}'")
        else:
            new = 'KABUPATEN ' + s
            logs.append(f"Row {idx}: '{s}' -> '{new}'")
        new_vals.append(new)

    DataDF['Kab/Kota'] = new_vals
    print("\n".join(logs))
```

```python

# Check null/invalid rows in both DataFrames, then print full entry + index
import pandas as pd

INVALID_TOKENS = {"", "na", "n/a", "null", "none", "-"}

def find_invalid_rows(df: pd.DataFrame, df_name: str):
    null_mask = df.isna()

    # Detect invalid text values (empty/placeholder strings)
    text_mask = (
        df.astype("string")
          .apply(lambda col: col.str.strip().str.lower().isin(INVALID_TOKENS))
          .fillna(False)
    )

    invalid_mask = null_mask | text_mask
    bad_rows = df[invalid_mask.any(axis=1)]

    print(f"\n=== {df_name} ===")
    if bad_rows.empty:
        print("No null/invalid values found.")
    else:
        print(f"Found {len(bad_rows)} row(s) with null/invalid values:")
        print(bad_rows)  # full rows, index shown on the left

find_invalid_rows(DataDF, "DataDF")
find_invalid_rows(CityDF, "CityDF")


```

```python
import pandas as pd

INVALID_TOKENS = {"", "na", "n/a", "null", "none", "-"}

def replace_invalid_with_dash(df: pd.DataFrame, df_name: str):
    logs = []
    affected_entries = 0
    affected_rows = set()

    for col in df.columns:
        s = df[col]

        null_mask = s.isna()

        # invalid text tokens (case-insensitive, trimmed)
        text_mask = (
            s.astype("string")
             .str.strip()
             .str.lower()
             .isin(INVALID_TOKENS)
             .fillna(False)
        )

        invalid_mask = null_mask | text_mask

        for idx in df.index[invalid_mask]:
            old_val = df.at[idx, col]

            # skip if already exactly "-"
            if (not pd.isna(old_val)) and (str(old_val) == "-"):
                continue

            df.at[idx, col] = "-"
            logs.append(f"[{df_name}] row={idx}, col='{col}': {repr(old_val)} -> '-'")
            affected_entries += 1
            affected_rows.add(idx)

    print(f"\n=== {df_name} CLEANING LOG ===")
    if logs:
        print("\n".join(logs))
    else:
        print("No changes made.")

    print(f"{df_name} affected rows: {len(affected_rows)}")
    print(f"{df_name} affected entries: {affected_entries}")
    return len(affected_rows), affected_entries


rows_data, entries_data = replace_invalid_with_dash(DataDF, "DataDF")
rows_city, entries_city = replace_invalid_with_dash(CityDF, "CityDF")

print("\n=== TOTAL SUMMARY ===")
print(f"Total affected rows (DataDF + CityDF): {rows_data + rows_city}")
print(f"Total affected entries (DataDF + CityDF): {entries_data + entries_city}")
```

```python
import pandas as pd

# --- Build CityCode / RegionalCode mapping from CityDF ---
city_map = (
    CityDF[["City", "CityCode", "RegionalCode"]]
    .drop_duplicates()
    .copy()
)

city_map["City_key"] = city_map["City"].astype("string").str.strip().str.upper()

# --- Normalize Kab/Kota for matching ---
df = DataDF.copy()
df["Kab_key"] = df["Kab/Kota"].astype("string").str.strip().str.upper()

# --- Join city codes to the whole rows of DataDF ---
df = df.merge(
    city_map[["City_key", "CityCode", "RegionalCode"]],
    left_on="Kab_key",
    right_on="City_key",
    how="left",
)

# --- Sort whole rows by CityCode, then Funcloc ascending inside each CityCode group ---
# Stable sort keeps row integrity and consistent ordering
df = df.sort_values(
    by=["CityCode", "Funcloc"],
    ascending=[True, True],
    kind="mergesort",
    na_position="last",
).reset_index(drop=True)

# --- Create CCC numbering starting from 001 within each CityCode group ---
df["CCC"] = df.groupby("CityCode", dropna=False).cumcount() + 1
df["CCC"] = df["CCC"].astype(str).str.zfill(3)

# --- Build Internal Site ID in the format AAA-BB-CCC ---
df["Internal Site ID"] = (
    df["CityCode"].astype("string").fillna("-") + "-" +
    df["RegionalCode"].astype("string").fillna("-") + "-" +
    df["CCC"]
)

# --- Drop unwanted columns ---
drop_cols = ["Alamat1", "Alamat2", "Alamat3", "Alamat4", "SiteName", "CityCode", "RegionalCode", "City_key", "Kab_key", "CCC"]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# --- Final column order ---
final_cols = ["Internal Site ID", "Kab/Kota"]
if "Funcloc" in df.columns:
    final_cols.append("Funcloc")

other_cols = [c for c in df.columns if c not in final_cols]
df = df[final_cols + other_cols]

# --- Update DataDF ---
DataDF = df

print("=== DataDF RESTRUCTURED ===")
print(DataDF.head(20))
print(f"\nDataFrame shape: {DataDF.shape}")
print(f"\nColumns: {DataDF.columns.tolist()}")
```
